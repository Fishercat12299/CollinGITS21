```python
city=['Seattle', 'San Francisco', 'Los Angeles', 'Denver', 'Salt Lake City', 'Phoenix', 'Chicago', 'St.Louis', 'Dallas', 'Boston', 'Washington DC', 'Miami']
totaldays=[306, 314, 275, 366, 366, 366, 366, 366, 306, 306, 366, 275]
poordays=[14, 22, 119, 34, 21, 70, 28, 10, 24, 0, 3, 2]
aqimed=[42, 45, 81, 58, 50, 75, 56, 52, 49, 42, 42, 44]
zone=['west','west','west', 'mountain', 'mountain', 'mountain', 'central', 'central', 'central', 'east', 'east', 'east']
```

```python
import pandas as pd
```

```python
aqicities = pd.DataFrame({'City':city,'TotalDays':totaldays, 'PoorDays':poordays, 'AQImedian':aqimed, 'Zone':zone})
```

```python
aqicities
```

```python
aqicities2 = pd.read_csv('aqicities2.csv')
```

```python
allcities = pd.concat([aqicities, aqicities2], axis=0, join='inner')
```
