```python
import pandas as pd
city = ['Seattle', 'San Francisco', 'Los Angeles', 'Denver', 'Salt Lake City', 'Phoenix', 'Chicago', 'St.Louis', 'Dallas', 'Boston', 'Washington DC', 'Miami', 'Clearlake', 'Salem', 'Durango', 'Sante Fe', 'Ann Arbor', 'Bismarck', 'Bangor', 'Charlottesville']
Totaldays2010 = [365, 365, 365, 365, 365, 365, 365, 365, 365, 365, 365, 365, 361, 365, 365, 365, 362, 365, 365, 261]
Poordays2010 = [3, 14, 115, 38, 24, 59, 43, 103, 28, 10, 50, 6, 0, 1, 10, 1, 1, 1, 0, 4]
AQImedian2010 = [41, 48, 77, 61, 51, 74, 71, 74, 52, 47, 55, 48, 35, 27, 48, 44, 38, 41, 34, 45]
zone = ['west', 'west', 'west', 'mountain', 'mountain', 'mountain', 'central', 'central', 'central', 'east', 'east', 'east', 'west', 'west', 'mountains', 'mountains', 'central', 'central', 'east', 'east']
```

```python
data2010 = pd.DataFrame({'City':city, 'TotalDays2010':Totaldays2010, 'PoorDays2010':Poordays2010, 'AQImedian2010':AQImedian2010, 'Zone':zone})
data2010
```

```python
data2020 = pd.read_csv('allcitiesAQI.csv')
data2020
```

```python
AllData = pd.merge(data2010,data2020,how='inner',on='City')
AllData
```

```python
AllData = AllData.drop(['Zone_x'], axis=1)
```

```python
AllData
```

```python
AllData = AllData.rename(columns={'Zone_y':'Zone'})
AllData
```

```python
AllData['Scale2010'] = AllData['TotalDays2010']/365
AllData['Scale2020'] = AllData['TotalDays2020']/365
AllData
```

```python
AllData['Poorday_Rate2010'] = AllData['PoorDays2010']/AllData['Scale2010']
AllData['Poorday_Rate2020'] = AllData['PoorDays2020']/AllData['Scale2020']
AllData['Poorday_Rate2010'] = AllData['Poorday_Rate2010'].astype(int)
AllData['Poorday_Rate2020'] = AllData['Poorday_Rate2020'].astype(int)
AllData
```

```python
AllData = AllData.drop(['Scale2010', 'Scale2020'], axis=1)
AllData
```

```python
AllData.plot.bar(x='City', y=['Poorday_Rate2010', 'Poorday_Rate2020'], ylabel = 'Rate of Poordays', title = 'Rate of Poor Days', width =.5)
```

```python
AllData.plot.bar(x='City', y=['AQImedian2010', 'AQImedian2020'], ylabel='AQImediann', width =.5, title = 'AQImedian Score for US Cities')
```

```python
Zone = AllData.groupby(['Zone']).mean()
Zone
```

```python
Zone.plot.bar(y=['PoorDays2010', 'PoorDays2010'], ylabel = 'Average', title = 'Average Number of Poor Days in Each Geographic Zone', width =.5)
```
